```python
import torch
from torch import Tensor
import dlc_practical_prologue as prologue

train_input, train_target, test_input, test_target = prologue.load_data(cifar = None, one_hot_labels = False, normalize = False, flatten = True)



```

```python
def nearest_classification(train_input, train_target, x):
    distVec = (train_input - x).pow(2).sum(1).view(-1)
    minDist,indexMin = torch.min(distVec,0)
    
    return train_target[indexMin[0]]

a = nearest_classification(train_input,train_target,test_input)

a
```
